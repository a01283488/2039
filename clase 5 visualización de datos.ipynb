```python
path = "Cartwheeldata.csv"
df = pd.read_csv(path)
df.head(2)
```

```python
import seaborn as sns

sns.lmplot(x = "Wingspan", # es L mplot (minuscula, no es i)
          y = "CWDistance",
          data = df,
          fit_reg = False, #Linea de regresion con true, sin linea con false
        hue = 'Gender') 

plt.title("tabla")
plt.show()
```

```python
sns.swarmplot (x='Gender',
             y= 'CWDistance',
             data = df,
             )
plt.title('Gender vs CSDistance')
plt.show()
```

```python
sns.boxplot(data=df.loc[df['Gender']=='M',
           ['Height','Wingspan','CWDistance']]
           )
plt.title('box plot')
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,3,1) #1 renglon, 2 cplummas, 1 posicion 1
sns.boxplot(data=df.loc[df['Gender']=='M',
                        ['Score']
                       ]
           )

ax2 = fig.add_subplot(1,3,3) 
sns.boxplot(data=df.loc[df['Gender']=='F',
                        ['Score']
                       ]
           )

ax1.title.set_text('Male Score')
ax2.title.set_text('Female Score')


plt.show()
```

```python
sns.distplot(df.CWDistance)
plt.show()
```

```python
sns.countplot(x='Gender',
             data=df)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
