```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('NEWYOKRDATA.csv')
df.head(2)
```

```python
#%ls
```

```python
df.columns
```

```python
data = df.sample (n=100,
                 random_state=20)
```

```python
line_data=data.groupby('CUISINE DESCRIPTION')['SCORE'].mean().reset_index()
line_data
```

```python
fig=go.Figure(data=go.Scatter(x=line_data['CUISINE DESCRIPTION'],
                             y=line_data['SCORE'],
                             mode = 'lines',
                             marker = dict(color='green')
                             )
             )
fig.show()
```

```python
fig=px.bar(line_data,
          x='CUISINE DESCRIPTION',
          y='SCORE',
          title='Bar chart')
fig.show()
```

```python
#regresio multiple 

import statsmodels.api as sm
model = sm.OLS.from_formula('BIN~SCORE', 
                           data = df)
result = model.fit()
result.summary()
```
