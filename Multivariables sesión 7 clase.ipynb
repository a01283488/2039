```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set()
```

```python
r=1
mean=[15,5]
cov = [[r,1],[r,1]]
x,y = x,y = np.random.multivariate_normal(mean,cov,400).T
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(x=x,
        bins=10)
plt.title('x')

plt.subplot(1,2,2)
plt.hist(x=y,
        bins=15)
plt.title('y')

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.hist(x=x,
        bins=10)
plt.title('x')

plt.subplot(2,2,2)
plt.hist(x=y,
        bins=15,
        orientation = 'horizontal')
plt.title('y')

plt.subplot(2,2,4)
plt.scatter(x=x,
           y=y
           )
plt.title('Scatter')

plt.show()
```

```python
df=pd.read_excel('session01/datasets/Canada.xlsx',
                sheet_name='Canada by Citizenship',
                skiprows = range(20),
                skipfooter = 2,
                engine = 'openpyxl'
                )
df.head(2)
```

```python
df.columns
```

```python
df.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'],
       axis=1, #se especifica de donde se eliminan
        inplace = True)
```

```python
df.columns
```

```python
df.rename(columns={
    'OdName':'Country',    
    'AreaName':'Continent',     
    'RegName': 'Region'
},
         inplace = True)
df.columns
```

```python
df.head(2)
```

```python
df.set_index('Country', 
            inplace = True)
```

```python
df.head(2)
```

```python
df.columns=list(map(str, df.columns))
df.columns
```

```python
df.head(1)
```

```python
df['Total']= df.iloc[:,3:].sum(axis=1)
df.head(2)
```

```python
df.iloc[:,3:4]
```

```python
years = list(map(str,range(1980,2014)))
years
```

```python
df_continentes = df.groupby('Continent',
                           axis = 0, #0 para columna, 1 para renglon
                           ).sum()
```

```python
df_continentes
```

```python
df_continentes.sort_values(['Total'],
                          ascending = False,
                          axis = 0,
                          inplace = True)
df_continentes.head()
```

```python
color_list =['skyblue','gold','yellowgreen', 'lightcoral','lightskyblue','lightgreen','pink']
explode_list=[0.1,0,0,0,0.2,0.3,0.4]

df_continentes['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct = '%10.1f%%', #para agregar porcentajes en el pie chart
                            startangle = 90,
                            shadow = False,
                             colors = color_list,
                             explode = explode_list)
plt.title ('Inmigration to Canada by Continent 1980-2013', y=1.2)
plt.axis('equal') #parece un circulo 

plt.legend(labels=df_continentes.index, loc='upper left')
plt.show()
```
